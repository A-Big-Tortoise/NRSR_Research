```python
def band_filter(signal, lowcut=0.1, highcut=8, order=4):

  # ba
  # 3
  # 
  b, a = butter(order, [lowcut, highcut], btype='band', fs=fs)

  # signal.lfilter
  # lfilter
  # ba
  # lfilter
  filtered_signal = lfilter(b, a, signal)

  return filtered_signal
```

```python
from PyEMD import EEMD

def standize(data):
    return (data - np.mean(data))/np.std(data)

def eemd_decomposition(signal, noise_width, ensemble_size=100, random_seed=None):
    if random_seed is not None:
        np.random.seed(random_seed)

    eemd = EEMD(trials=ensemble_size, noise_width=noise_width)
    imfs = eemd.eemd(signal)
    print(imfs.shape)

    return imfs


show = True
all_imfs = []

for cnt, o_signal in enumerate(signals_clean):

  signal = band_filter(standize(o_signal))
  # signal = standize(band_filter(o_signal))

  imfs = eemd_decomposition(signal, noise_width=0.05)
  all_imfs.append(imfs)

  if show:
    num_imfs = len(imfs)
    t = np.linspace(0, duration, duration * fs)

    plt.figure(figsize=(10, 6))
    plt.subplot(num_imfs + 1, 1, 1)
    plt.plot(t, signal, 'b', label='Original Signal')
    plt.legend()

    for i, imf in enumerate(imfs):
        plt.subplot(num_imfs + 1, 1, i + 2)
        plt.plot(t, imf, 'r', label='IMF {}'.format(i + 1))
        plt.legend()

    plt.tight_layout()
    plt.show()
    show = False
  # break
  if cnt > 100:
    break

print(len(all_imfs))
```

```python
def calculate_frequency_mean(signal, sample_rate, show=False):
    N = len(signal)
    frequencies = np.fft.fftfreq(N, d=1/sample_rate)
    fft_result = np.fft.fft(signal)

    positive_frequencies = frequencies[:N//2]
    positive_fft_result = fft_result[:N//2]

    # weighted_sum = np.sum(positive_frequencies * np.abs(positive_fft_result))
    # weighted_sum = np.sum(positive_frequencies * positive_fft_result)
    # total_weight = np.sum(np.abs(positive_fft_result))
    freq = np.argmax(np.abs(positive_fft_result))

    # print('argmax_freq: ', freq / ((N // 2)//50))
    if show:
      plt.figure(figsize=(10, 4))
      plt.plot(positive_frequencies, np.abs(positive_fft_result))
      plt.title('Frequency Spectrum')
      plt.xlabel('Frequency (Hz)')
      plt.ylabel('Amplitude')
      plt.grid(True)
      plt.xlim(0, 10)  # 
      plt.show()

    # if total_weight == 0:
    #     return 0
    # else:
    #     frequency_mean = weighted_sum / total_weight
    #     return frequency_mean
    if freq == 0:
      freq = 1

    return freq / ((N // 2)//50)

all_freqs = []
for imfs in all_imfs:
  freqs = []
  for imf in imfs:
    argmax_frequency = calculate_frequency_mean(imf, sampling_rate)
    # if
    # print('Argmax frequency:', argmax_frequency)
    freqs.append(argmax_frequency)

  all_freqs.append(freqs)

  # print()
  # break
```

```python
from sklearn.decomposition import PCA

def get_most_important_component(data_matrix, num_components=1):
    """
    PCA

    
    data_matrix (array-like): 
    num_components (int): 1

    
    most_important_component (array): 
    """
    # PCA
    pca = PCA(n_components=num_components)

    # PCA
    pca.fit(data_matrix)
    transformed_data = pca.transform(data_matrix)

    # 
    # print( pca.components_)
    most_important_component = pca.components_[0]

    return most_important_component

hr_low_freq, hr_high_freq = 0.75, 2.51
rr_low_freq, rr_high_freq = 0.1, 0.7

hr_real, hr_estimated_S1, hr_estimated_S2, hr_estimated_S3 = [], [], [], []
rr_real, rr_estimated_S1, rr_estimated_S2, rr_estimated_S3 = [], [], [], []

for cnt, (freqs, imfs) in enumerate(zip(all_freqs, all_imfs)):
  freqs = np.array(freqs)
  positions_hr = np.bitwise_and(freqs > hr_low_freq, freqs < hr_high_freq)
  positions_rr = np.bitwise_and(freqs > rr_low_freq, freqs < rr_high_freq)
  imfs_hr, imfs_rr = imfs[positions_hr], imfs[positions_rr]
  # print(positions_hr, positions_rr)
  # print(imfs_hr.shape, imfs_rr.shape)

  # S 1

  if len(imfs_hr) == 1:
    hr_most_important_component_S1 = imfs_hr[0]
  else:
    hr_most_important_component_S1 = get_most_important_component(imfs_hr)
  if len(imfs_rr) == 1:
    rr_most_important_component_S1 = imfs_rr[0]
  else:
    rr_most_important_component_S1 = get_most_important_component(imfs_rr)

  # S 2
  hr_most_important_component_S2 = get_most_important_component(imfs_hr)
  rr_most_important_component_S2 = get_most_important_component(imfs_rr)

  # S 3
  hr_most_important_component_S3 = imfs_hr.sum(0)
  rr_most_important_component_S3 = imfs_rr.sum(0)

  freq_hr_S1, freq_rr_S1 = calculate_frequency_mean(hr_most_important_component_S1, 100), calculate_frequency_mean(rr_most_important_component_S1, 100)
  freq_hr_S2, freq_rr_S2 = calculate_frequency_mean(hr_most_important_component_S2, 100), calculate_frequency_mean(rr_most_important_component_S2, 100)
  freq_hr_S3, freq_rr_S3 = calculate_frequency_mean(hr_most_important_component_S3, 100), calculate_frequency_mean(rr_most_important_component_S3, 100)

  # print(freq_hr, freq_rr)
  # print(freq_hr * 60, freq_rr * 60)

  # print(labels[cnt][2], labels[cnt][3])

  hr_real.append(labels[cnt][2])

  hr_estimated_S1.append(freq_hr_S1 * 60)
  hr_estimated_S2.append(freq_hr_S2 * 60)
  hr_estimated_S3.append(freq_hr_S3 * 60)

  rr_real.append(labels[cnt][3])
  rr_estimated_S1.append(freq_rr_S1 * 60)
  rr_estimated_S2.append(freq_rr_S2 * 60)
  rr_estimated_S3.append(freq_rr_S3 * 60)

  # print()
  # break
```
